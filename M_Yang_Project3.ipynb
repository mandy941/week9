```python
# import csv
# define numpy as np
# define panda as pd
# define re
# read the txt file from my own path
# create a empty list
# use for loop to generate a lists that contains multipe lists, each list represent each rows of the txt file
#print the results
import csv 
import numpy as np
import pandas as pd
import re
with open (r'C:\Users\13750\Desktop\yu\DAV5400 Analytics Programming\Assignment\week9-project3\tournamentinfo.txt') as txtfile:
    readCSV = csv.reader(txtfile)
    table = []
    for row in readCSV:
        table.append(row)
table
```

```python
# extract the lists after the third list, the rows after the forth can seem like a regular formulas
# print the results
table1 = table[3:]
table1
```

```python
# use len function to get the length of table1
# make row begins as 0
# create an empty list
# use the for loop to get the lists, begins as second list 
# print the results
length = len(table1)
b = np.arange(1,length,3)
row = 0
table2 = []
for a in table1:
    if (row in b):
        table2.append(a)
    row = row+1
table2
```

```python

# make row begins as 0
# create an empty list
# use the for loop to get the lists , begins as third list 
# print the results
c = np.arange(2,length,3)
row = 0
table3 = []
for a in table1:
    if (row in c):
        table3.append(a)
    row = row+1
table3
```

```python
# define re
# use compile function and regex to get parts 
# use findall function to find the specific string in table2 list by regex
# print the result
import re
str1 = re.compile('[A-Z]+\s[A-Z]+\s[A-Z]*')
name = str1.findall(str(table2))
name
```

```python
# use compile function and regex to get parts 
# use findall function to find the specific string in table3 list by regex
#print the result
str2 = re.compile('[A-Z]{2}')
state = str2.findall(str(table3))
state
```

```python
# use compile function and regex to get parts 
# use findall function to find the specific string in table2 list by regex
#print the result
str3 = re.compile('[0-9]\.[0-9]')
rate = str3.findall(str(table2))
rate
```

```python
# use compile function and regex to get parts 
# use findall function to find the specific string in table3 list by regex
#print the result
str4 = re.compile('[A-Z]{1}\:\s+[0-9]+')
pre1 = str4.findall(str(table3))
pre1
```

```python
# use compile function and regex to get the number 
# use findall function to find the specific regex in pre1 list
#print the result
str5 = re.compile('[0-9]+')
pre2 = str5.findall(str(pre1))
#print the result
pre2
```
