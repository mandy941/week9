```python
# import csv
# define numpy as np
# define panda as pd
# print the result as df
import csv 
import numpy as np
import pandas as pd
with open (r'C:\Users\13750\Desktop\yu\DAV5400 Analytics Programming\Assignment\week9-project3\tournamentinfo.txt') as txtfile:
    readCSV = csv.reader(txtfile)
    table = []
    for row in readCSV:
        table.append(row)
table
```

```python
#remove the first three unmeaningful rows, the rows after the forth can seem like a regular formulas
table1 = table[3:]
table1
```

```python
length = len(table1)
b = np.arange(1,length,3)
row = 0
table2 = []
for a in table1:
    if (row in b):
        table2.append(a)
    row = row+1
table2
```
