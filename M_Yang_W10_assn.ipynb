```python
# define panda, numpy, re
import numpy as np
import pandas as pd
import re
# read the data file from GitHub as row style
# print the result
df = pd.read_csv('https://raw.githubusercontent.com/mandy941/week9/master/pre.csv')
df
```

# Task one : converted to “long” format
1. drop the empty column
2. rename the unnamed column
3. fill the empty box
4. convert the dataframe to be long format
5. rename the variable and value column
6. print the result

```python
# use drop function move out the third column whose index is 2
#print the result
df1 = df.drop(df.index[2])
df1
```

```python
#use rename function to rename the unname column 
#print the result
df2 = df1.rename(columns={'Unnamed: 0':'Airline','Unnamed: 1':'Status'})
df2
```

```python
# fill the empty box by loc function
# print the result
df2.loc[1,'Airline'] = df2.loc[0,'Airline']
df2.loc[4,'Airline'] = df2.loc[3,'Airline']
df2
```

```python
#use melt function to make table df2 to be long format
#print the result
long = df2.melt(id_vars = ['Airline', 'Status'], value_vars = ['Los Angeles','Phoenix','San Diego','San Francisco','Seattle'])
long
```

```python
#rename the column variable and value
long1 = long.rename ( columns = { 'variable' : 'Destination', 'value' : 'Times'})
long1
```

# Task 3: perform analysis to compare the arrival delays for the two airlines
1. make ascending order for Airline column 
2. create a dataframe that show the status is on time flights
3. drop the unmeaningful index column
4. extract the data that the status is delay 
4. group the Airline and sum the Times column 
5. group the Airline,Destination and sum the Times column 
6. get the total delay rate for two airlines
7. make a conclusion
8. get the delay rate of two airlines for each city
9. make reault table from long format to wide format
10. make a conclusion

```python
# use sort_values funtion to make ascending order for Airline column
# use rest_index funtion to renew the index for the new table
# print the result
A = long1.sort_values(['Airline','Status'], ascending=[1,0]).reset_index()
A = A.drop(['index'],axis = 1)
A
```

```python
#use rest_index funtion to renew the index for new table
# drop the old index column
# print the result
delay = A[A['Status'] == 'delayed'].reset_index()
delay = delay.drop(['index'],axis = 1)
delay
```

```python
# use to_numeric function to make column Times from string to numeric
# use groupby funtion to find the total delay times of two airlines 
# print the result
delay['Times'] = pd.to_numeric(delay['Times'])
delayed = delay.groupby(['Airline'])[['Times']].sum()
delayed
```

```python
# use groupby funtion to find the total operating times of two airlines 
# print the result
total = A.groupby(['Airline'])[['Times']].sum()
total
```

```python
# use delayed table to divide the total table to get the delay rate.
# print the result
delay_rate = delayed['Times']/total['Times']
delay_rate
```

According to the delay rate of two airlines, we can know AM WEST delay rate is lower than ALASKA airlines. It means that AM WEST has better performance for all the works.

```python
# use groupby function and rest_index funtion to find the total operating times of tow airlines for each city
# print the result
total1 = A.groupby(['Airline','Destination'])[['Times']].sum().reset_index()
total1
```

```python
# make delay times of tow airlines for each city divide by totoal delay times of tow airlines for each city
# print the result
total1['citydelay'] = delay['Times']/total1['Times']
total1
```

```python
# use pivot_table funtion to change the table to be the wide format
# print the result
wide = pd.pivot_table(total1, values = 'citydelay', index = ['Airline'],columns = ['Destination'])
wide
```

According to the wide table, we can know the delay rate of two airlines for each city, the lower delay rate means the better performance. Therefore, for each city, the airlines has better performance is as following.
